```python
from Bio import Entrez
import pandas as pd

Entrez.email = 'A.N.Other@example.com'

def query_NCBI(gene_ID):
    handle = Entrez.efetch(db="gene", id=gene_ID, rettype="gb", retmode="text")
    # print(handle.url)
    lines = handle.readlines()
    organism = lines[2].strip().split('[')[1].strip(']')
    if organism == "Saccharomyces cerevisiae S288C" or organism == "Schizosaccharomyces pombe (fission yeast)":
        geneSymbol = lines[3].strip().split(':')[1] # .split(', ')
    else:
        geneSymbol = lines[2].strip().split(' ')[2]
    # geneID = lines[8].strip().split(':')[1]
    handle.close()
    return [int(gene_ID), geneSymbol, organism]
```

```python
NCBI_gene_IDs = pd.read_csv('ModuLand_clusters_NCBI.csv')
NCBI_genes = NCBI_gene_IDs.applymap(lambda x: query_NCBI(x), na_action="ignore")
```

```python
NCBI_genes_hs = NCBI_genes.iloc[[0,2,3,4,6,9,10,14,19,22]].reset_index(drop=True)
NCBI_genes_hs_dict = NCBI_genes_hs.to_dict()
NCBI_genes_hs_dict

with open('NCBI_human_genes.csv', 'w') as fh:
    fh.write('ncbi_gene_id,gene_symbol\n')
    for col, rows in NCBI_genes_hs_dict.items():
        for row,value in rows.items():
            if isinstance(value, list):
                NCBI = str(value[0])
                symbol = value[1]
                line = ','.join([NCBI,symbol])
                fh.write(line + '\n')
```

```python
NCBI_genes_other = NCBI_genes.iloc[[1,5,7,8,11,12,13,15,16,17,18,20,21,23,24,25,26]].reset_index(drop=True)
NCBI_genes_other_dict = NCBI_genes_other.to_dict()
NCBI_genes_other_dict

with open('NCBI_other_genes.csv', 'w') as fh:
    fh.write('ncbi_gene_id,organism,gene_symbol\n')
    for col, rows in NCBI_genes_other_dict.items():
        for row,value in rows.items():
            if isinstance(value, list):
                NCBI = str(value[0])
                symbol = value[1]
                organism = value[2]
                line = ','.join([NCBI,organism,symbol])
                fh.write(line + '\n')
```

```python
gene_IDs = pd.read_csv('genes.csv')
gene_IDs = gene_IDs.drop_duplicates(ignore_index=True)
gene_ID_list = gene_IDs['ncbi_gene_id'].to_list()

# NCBI_gene_IDs = pd.DataFrame(columns=['ncbi_gene_id', 'organism','geneSymbol'])
def queryNCBI(gene_ID):
    handle = Entrez.efetch(db="gene", id=gene_ID, rettype="gb", retmode="text")
    # print(handle.url)
    lines = handle.readlines()
    organism = lines[2].strip().split('[')[1].strip(']')
    if organism == "Saccharomyces cerevisiae S288C" or organism == "Schizosaccharomyces pombe (fission yeast)":
        geneSymbol = lines[3].strip().split(':')[1] # .split(', ')
    else:
        geneSymbol = lines[2].strip().split(' ')[2]
    # geneID = lines[8].strip().split(':')[1]
    handle.close()
    return int(gene_ID), geneSymbol, organism

NCBI_dict = {
    'ncbi_gene_id': [],
    'organism': [],
    'geneSymbol': []
}
for gene in gene_ID_list:
    gene_ID, geneSymbol, organism = queryNCBI(gene)
    NCBI_dict['ncbi_gene_id'].append(gene_ID)
    NCBI_dict['organism'].append(organism)
    NCBI_dict['geneSymbol'].append(geneSymbol)

NCBI_gene_IDs = pd.DataFrame.from_dict(NCBI_dict)
```
