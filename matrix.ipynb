```python
import requests
import pandas as pd
from requests.packages.urllib3.exceptions import InsecureRequestWarning
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
genes = pd.read_csv('NCBI_genes.csv')
genes = genes.drop_duplicates(ignore_index=True)
phenotypes = pd.read_csv('longQT_phenotypes.csv')
```

```python
base_url = 'https://hpo.jax.org/api/hpo/term/'
genes_url = '/genes'
data = []
for i in range(len(phenotypes)):
    url = base_url + phenotypes.iloc[i,0] + genes_url+"?max=-1"
    response = requests.get(url, verify=False)
    data_dict = response.json()
    phenotype_genes = [item['entrezGeneId'] for item in data_dict['genes']]
    data.append(list(map(int, genes['ncbi_gene_id'].isin(phenotype_genes))))
matrix = pd.DataFrame(data, index=phenotypes, columns=genes).transpose()
# remove rows of all 0s
matrix = matrix.loc[~(matrix==0).all(axis=1)]
adjacency_matrix = matrix.dot(matrix.transpose())
```

```python
fig, ax = plt.subplots(figsize = (21, 14))
sns.heatmap(matrix, cmap ='RdYlBu', linewidths = 0.30, square=False)
```

```python
fig, ax = plt.subplots(figsize = (21, 14))
sns.heatmap(adjacency_matrix, cmap ='viridis', linewidths = 0.30)
```

```python
sns.set(font_scale=1)
sns.clustermap(adjacency_matrix, cmap ='viridis', linewidths = 0.30, figsize=(20,20))
```
